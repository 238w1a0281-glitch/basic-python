```python
cart_items = []
total_bill = 0

def tiffen_items():
    global total_bill
    print("Welcome to Tiffen Section \nWhat type of Tiffen you want like \n1.Dosa \n2.Idly \n3.Puri \n4.Bajji")
    a = int(input("Enter the Tiffen you would like in (1-4): "))
    if a == 1:
        price = 25
        item = "Dosa"
    elif a == 2:
        price = 20
        item = "Idly"
    elif a == 3:
        price = 15
        item = "Puri"
    elif a == 4:
        price = 10
        item = "Bajji"
    else:
        print("Invalid input")
        return

    b = int(input("Enter the quantity you would like: "))
    bill = b * price
    total_bill += bill
    cart_items.append(f"{item} x{b} = {bill}")
    print("Total price is", bill)

def meals():
    global total_bill
    print("Welcome to Meals Section \nWhat type of Meals you want like \n1.Dum Biriyani  \n2.Veg Biriyani \n3.Full Veg meals \n4.Full Non Veg meals")
    a = int(input("Enter the Meals you would like in (1-4): "))
    if a == 1:
        price = 100
        item = "Dum Biriyani"
    elif a == 2:
        price = 80
        item = "Veg Biriyani"
    elif a == 3:
        price = 120
        item = "Full Veg meals"
    elif a == 4:
        price = 150
        item = "Full Non Veg meals"
    else:
        print("Invalid input")
        return

    b = int(input("Enter the quantity you would like: "))
    bill = b * price
    total_bill += bill
    cart_items.append(f"{item} x{b} = {bill}")
    print("Total price is", bill)

def Juice():
    global total_bill
    print("Welcome to juice Section \nWhat type of juice you want like \n1.Pomegranate  \n2.Apple \n3.Banana \n4.Mango")
    a = int(input("Enter the juice you would like in (1-4): "))
    if a == 1:
        price = 70
        item = "Pomegranate Juice"
    elif a == 2:
        price = 60
        item = "Apple Juice"
    elif a == 3:
        price = 40
        item = "Banana Juice"
    elif a == 4:
        price = 50
        item = "Mango Juice"
    else:
        print("Invalid input")
        return

    b = int(input("Enter the quantity you would like: "))
    bill = b * price
    total_bill += bill
    cart_items.append(f"{item} x{b} = {bill}")
    print("Total price is", bill)

def snacks():
    global total_bill
    print("Welcome to snacks Section \nWhat type of snacks you want like \n1.Pizza  \n2.Burger \n3.Noodles \n4.Puff")
    a = int(input("Enter the snacks you would like in (1-4): "))
    if a == 1:
        price = 100
        item = "Pizza"
    elif a == 2:
        price = 80
        item = "Burger"
    elif a == 3:
        price = 120
        item = "Noodles"
    elif a == 4:
        price = 150
        item = "Puff"
    else:
        print("Invalid input")
        return

    b = int(input("Enter the quantity you would like: "))
    bill = b * price
    total_bill += bill
    cart_items.append(f"{item} x{b} = {bill}")
    print("Total price is", bill)

def all():
    tiffen_items()
    meals()
    Juice()
    snacks()

def show_cart():
    global total_bill
    print("\nYour Cart:")
    if cart_items:
        for item in cart_items:
            print(item)
        print(f"Total Bill: {total_bill}")
    else:
        print("Your cart is empty!")

while True:
    print("\nWelcome to Food delivery system")
    print("Choose the Food item \n1.Tiffens \n2.Meals \n3.Juice \n4.Snacks \n5.To show all menu \n6.Cart \n7.Exit")
    a = int(input("Enter your choice (1-7): "))
    if a == 1:
        tiffen_items()
    elif a == 2:
        meals()
    elif a == 3:
        Juice()
    elif a == 4:
        snacks()
    elif a == 5:
        all()
    elif a == 6:
        show_cart()
    elif a == 7:
        print("Thank you for using the Food Delivery System!")
        break
    else:
        print("Invalid input, please try again.")
```
